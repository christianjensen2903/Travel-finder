```python
class Flight(BaseModel):
    departure: Airport
    arrival: Airport
    departure_time: datetime.datetime
    arrival_time: datetime.datetime
    price: float
    stops: int


def parse_flights(data: dict, from_airport: Airport, to_airport: Airport) -> list[Flight]:
    try:
        offers = data["data"]["flightOffers"]
        flights = []
        
        for offer in offers:
            departure_time = datetime.datetime.fromisoformat(offer["segments"][0]["departureTime"])
            arrival_time = datetime.datetime.fromisoformat(offer["segments"][-1]["arrivalTime"])
            flights.append(Flight(
                departure=from_airport,
                arrival=to_airport,
                departure_time=departure_time,
                arrival_time=arrival_time,
                price=offer["priceBreakdown"]["total"]["units"],
                stops=len(offer["segments"]) - 1,
            ))

        return flights
    except Exception as e:
        print(data)
        print(e)
        # raise e
```

```python
import requests

def parse_ryanair_flights(data: dict, from_airport: Airport, to_airport: Airport, adults: int = 2,) -> list[Flight]:
    try:
        offers = data["fares"]
        flights = []
        
        for offer in offers:
            departure_time = datetime.datetime.fromisoformat(offer["outbound"]["departureDate"])
            arrival_time = datetime.datetime.fromisoformat(offer["outbound"]["arrivalDate"])
            flights.append(Flight(
                departure=from_airport,
                arrival=to_airport,
                departure_time=departure_time,
                arrival_time=arrival_time,
                price=offer["outbound"]["price"]["value"] * adults,
                stops=0,
            ))

        return flights
    except Exception as e:
        print(data)
        print(e)
        # raise e

def get_ryanair_flights(origin: Airport, destination: Airport, date: datetime.date, adults: int = 2) -> list[Flight]:
    url = f"https://services-api.ryanair.com/farfnd/3/oneWayFares"

    # Parse date to 'YYYY-MM-DD' format
    date = date.strftime('%Y-%m-%d')

    query_params = {
        "departureAirportIataCode": origin.code,
        "arrivalAirportIataCode": destination.code,
        "outboundDepartureDateFrom": date,
        "outboundDepartureDateTo": date,
        "currency": "DKK",
    }

    response = requests.get(url, params=query_params)
    data = response.json()
    return parse_ryanair_flights(data, origin, destination, adults)

get_ryanair_flights(airports[1], cph_airport, datetime.date(2024, 9, 2))
```

```python
def search_flights(from_airport: Airport, to_airport: Airport, departure_date: datetime.date, adults: int = 2, currency_code: str = "DKK"):
    global api_calls
    url = "https://booking-com15.p.rapidapi.com/api/v1/flights/searchFlights"

    departDate_formatted = departure_date.strftime("%Y-%m-%d")

    querystring = {
        "fromId": f"{from_airport.code}.AIRPORT",
        "toId": f"{to_airport.code}.AIRPORT",
        "departDate": departDate_formatted,
        "pageNo": "1",
        "adults": adults,
        "sort": "BEST",
        "cabinClass": "ECONOMY",
        "currency_code": currency_code
    }

    headers = {
        "x-rapidapi-key": RAPIDAPI_KEY,
        "x-rapidapi-host": "booking-com15.p.rapidapi.com"
    }
    response = requests.get(url, headers=headers, params=querystring)
    data = response.json()
    flights = parse_flights(data, from_airport, to_airport)
    api_calls += 1
    ryan_air_flights = get_ryanair_flights(from_airport, to_airport, departure_date, adults) 
    return flights + ryan_air_flights
```

```python
def generate_dates(weeks=4):
    today = datetime.datetime.today()
    days = {"Thursday": 3, "Friday": 4, "Sunday": 6, "Monday": 0}  # Mapping days to weekday indices
    dates = {"Thursday": [], "Friday": [], "Sunday": [], "Monday": []}
    
    for i in range(1,weeks+1):
        for day, weekday in days.items():
            

            date = today + datetime.timedelta(days=(weekday - today.weekday() + 7 * i))
            if day == "Monday": # If the day is Monday, we need to add 7 days to the date
                date += datetime.timedelta(days=7)
            dates[day].append(date)  # Append the date in YYYY-MM-DD format

    return dates

generate_dates(2)
```

```python
import requests
import json

def get_city_id(city_name: str) -> str | None:
    global api_calls

    # Read city id from city_ids.json
    with open("city_ids.json", "r") as file:
        city_ids = json.load(file)
        if city_name in city_ids:
            return city_ids[city_name]

    url = "https://booking-com15.p.rapidapi.com/api/v1/hotels/searchDestination"

    querystring = {
        "query": city_name,
    }

    headers = {
        "x-rapidapi-key": RAPIDAPI_KEY,
        "x-rapidapi-host": "booking-com15.p.rapidapi.com"
    }
    response = requests.get(url, headers=headers, params=querystring)
    api_calls += 1
    data = response.json()["data"]
    for row in data:
        if row["search_type"] == "city":
            # Save city id to city_ids.json
            city_ids[city_name] = row["dest_id"]
            with open("city_ids.json", "w") as file:
                json.dump(city_ids, file)
                
            return row["dest_id"]
    return None
    

def get_hotel_price(city_name: str, checkin_date: datetime.date, checkout_date: datetime.date, adults: int = 2, currency_code: str = "DKK"):
    global api_calls
    """Get Q1 for the top-rated hotels in a city"""
    city_id = get_city_id(city_name)
    if city_id is None:
        return []
    

    url = "https://booking-com15.p.rapidapi.com/api/v1/hotels/searchHotels"

    checkinDate_formatted = checkin_date.strftime("%Y-%m-%d")
    checkoutDate_formatted = checkout_date.strftime("%Y-%m-%d")

    querystring = {
        "dest_id": city_id,
        "search_type": "CITY",
        "adults": adults,
        "departure_date": checkoutDate_formatted,
        "arrival_date": checkinDate_formatted,
        "currency_code": currency_code,
        "sort_by": "popularity"
    }

    headers = {
        "x-rapidapi-key": RAPIDAPI_KEY,
        "x-rapidapi-host": "booking-com15.p.rapidapi.com"
    }
    response = requests.get(url, headers=headers, params=querystring)
    api_calls += 1
    data = response.json()
    prices = []
    for hotel in data["data"]["hotels"]:
        price = hotel["property"]["priceBreakdown"]["grossPrice"]["value"]
        prices.append(price)

    # Find first quartile
    q1 = np.percentile(prices, 25)
        
    return q1

get_hotel_price("Paris", datetime.date(2024, 9, 27), datetime.date(2024, 9, 29))
```

```python
from tqdm import tqdm

weeks = 10
dates = generate_dates(weeks)

flights: list[Flight] = []
for airport in tqdm(airports):
    for departure_day in ["Thursday", "Friday", "Sunday", "Monday"]:
        for i in range(weeks): 

            if api_calls >= max_api_calls:
                break
            
            departure_date = dates[departure_day][i]

            if departure_day == "Thursday" or departure_day == "Friday":
                flights.extend(search_flights(from_airport=cph_airport, to_airport=airport, departure_date=departure_date))
            else:
                flights.extend(search_flights(from_airport=airport, to_airport=cph_airport, departure_date=departure_date))
```

```python
hotels = {}

for airport in tqdm(airports):

    if airport.city in hotels:
        continue
    else:
        hotels[airport.city] = []

    for i in range(weeks): 

        if api_calls >= max_api_calls:
            break
        
        for departure_day in ["Thursday", "Friday"]:
            departure_date = dates[departure_day][i]
            for arrival_day in ["Sunday", "Monday"]:
                arrival_date = dates[arrival_day][i]
                price = get_hotel_price(airport.city, departure_date, arrival_date)
                hotels[airport.city] += [{
                    "checkin_date": departure_date,
                    "checkout_date": arrival_date,
                    "price": price
                }]
```

```python
df = pd.json_normalize([flight.model_dump() for flight in flights])
```

```python
# Remove "bad" arrivals (arrivals between 22.00 and 06.00)
df = df[(df["arrival_time"].dt.hour >= 5) & (df["arrival_time"].dt.hour < 22)]

# Remove flights with more than 1 stop
# df = df[df["stops"] == 0]

# Remove flights longer than 5 hours
df["duration"] = df["arrival_time"] - df["departure_time"]
df = df[df["duration"] < pd.Timedelta(hours=5)]
```

```python
# Define country column (the column of arrival.country or departure.country which is not Denmark)
df["city"] = df.apply(lambda row: row["arrival.city"] if row["arrival.city"] != "Copenhagen" else row["departure.city"], axis=1)
```

```python
# Remove Beauvais-Tillé Airport
df = df[df["arrival.code"] != "BVA"]
df = df[df["departure.code"] != "BVA"]
```

```python
# Remove flights later than 2024-10-17
# Remove flights earlier than 2024-09-09
df = df[df["departure_time"] <= datetime.datetime(2024, 10, 17)]
df = df[df["departure_time"] >= datetime.datetime(2024, 9, 9)]
```

```python
# Step 1: Adjust week numbers for Monday flights (so they belong to the previous week)
df['week_number'] = df['departure_time'].dt.isocalendar().week
df.loc[df["departure_time"].dt.dayofweek == 0, 'week_number'] -= 1

# Step 2: Define the two weekend options (Thursday-Sunday and Friday-Monday)
# First option (Thursday-Sunday)
thursday_flights = df[(df["departure_time"].dt.dayofweek == 3) & (df["departure_time"].dt.hour >= 18)]
cheapest_thursday_flights = thursday_flights.loc[thursday_flights.groupby(['city', 'week_number'])["price"].idxmin()]

early_friday_flights = df[(df["arrival_time"].dt.dayofweek == 4) & (df["arrival_time"].dt.hour < 14)]
cheapest_early_friday_flights = early_friday_flights.loc[early_friday_flights.groupby(['city', 'week_number'])["price"].idxmin()]

sunday_flights = df[(df["departure_time"].dt.dayofweek == 6) & (df["departure_time"].dt.hour >= 14)]
cheapest_sunday_flights = sunday_flights.loc[sunday_flights.groupby(['city', 'week_number'])["price"].idxmin()]

# Combine the departure and return flights for the first option, ensuring they are from the same week
thursday_option = pd.merge(cheapest_thursday_flights, cheapest_sunday_flights, on=["city", "week_number"], suffixes=("_out", "_back"))
early_friday_option = pd.merge(cheapest_early_friday_flights, cheapest_sunday_flights, on=["city", "week_number"], suffixes=("_out", "_back"))

# Second option (Friday-Monday)
late_friday_flights = df[(df["departure_time"].dt.dayofweek == 4) & (df["departure_time"].dt.hour >= 18)]
cheapest_late_friday_flights = late_friday_flights.loc[late_friday_flights.groupby(['city', 'week_number'])["price"].idxmin()]

early_saturday_flights = df[(df["arrival_time"].dt.dayofweek == 5) & (df["arrival_time"].dt.hour < 14)]
cheapest_early_saturday_flights = early_saturday_flights.loc[early_saturday_flights.groupby(['city', 'week_number'])["price"].idxmin()]


monday_flights = df[(df["departure_time"].dt.dayofweek == 0) & (df["departure_time"].dt.hour >= 14)]
cheapest_flight_back_2 = monday_flights.loc[monday_flights.groupby(['city', 'week_number'])["price"].idxmin()]

late_friday_option = pd.merge(cheapest_late_friday_flights, monday_flights, on=["city", "week_number"], suffixes=("_out", "_back"))
saturday_option = pd.merge(cheapest_early_saturday_flights, monday_flights, on=["city", "week_number"], suffixes=("_out", "_back"))



# Step 3: Concatenate both options
all_options = pd.concat([thursday_option, early_friday_option, late_friday_option, saturday_option]).reset_index()

# Add hotel prices
def lookup_hotel_price(row):
    city = row["city"]
    departure_date = row["departure_time_out"].date()
    arrival_date = row["arrival_time_back"].date()
    for hotel in hotels[city]:
        if hotel["checkin_date"].date() == departure_date and hotel["checkout_date"].date() == arrival_date:
            return hotel["price"]
all_options["hotel_price"] = all_options.apply(lookup_hotel_price, axis=1)

# Calculate total price
all_options["total_price"] = all_options["price_out"] + all_options["price_back"] + all_options["hotel_price"]

# Step 4: Group by country and week, then keep the option with the lowest total price
cheapest_options = all_options.loc[all_options.groupby(['city'])['total_price'].idxmin()]

# Print the results
for index, row in cheapest_options.iterrows():
    print(f"City: {row['city']}")
    print(f"Week: {row['week_number']}")
    print(f"Departure: {row['departure_time_out']} - {row['arrival_time_out']}")
    print(f"Return: {row['departure_time_back']} - {row['arrival_time_back']}")
    print(f"Flight Price: {row['price_out']} + {row['price_back']}")
    print(f"Hotel Price: {row['hotel_price']}")
    print(f"Total Price: {row['total_price']}\n")
```
